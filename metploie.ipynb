```python
np.random.seed(5)
data = np.random.randn(100)
```

```python


plt.hist(data , bins= 10 , color='r' , alpha =0.7 , edgecolor='black' ,label='paul')
plt.grid()
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('basic historam ')
plt.legend()
plt.show()
```

```python
size = [40,30,20,10]
labels = ['group 1' , 'group 2' , 'group 3' , 'group 4']
#colors =['blue' , 'green' , 'red' ,'yellow']

plt.pie(size , labels=labels , autopct='%1.1f%%' , startangle=180 ,explode=(0.1 ,0.05, 0 , 0) )
plt.legend()
plt.title('Pie Chart')
plt.show()
```

```python
categories = ['a','b','c','d','e']
values = [5,7,3,8,4]
width= [ 0.10 , 0.15, 0.20, 0.30, 0.35]
color = ['blue' , 'red' , 'green' , 'purple' , 'yellow']
plt.bar(categories , values ,width=width , color=color , edgecolor='black')

plt.title('mid term marks ')
plt.grid(True)
plt.show()
```

```python
w = 0.2
x = ['c1' ,'c2' , 'c3','c4']
boys=[100 , 40 ,60 ,50]
girls = [200, 70, 30, 70]
bar1= np.arange(len(x))
plt.bar(bar1, boys,w,label='boys')
bar2 =[i+w for i in bar1]
plt.bar(bar2,girls,w,label ='girls')
plt.xticks(bar1+w/2,x)
plt.show()
```
